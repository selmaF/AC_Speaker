```python
import pandas as pd
import numpy as np
data=pd.read_csv('kp_coords/picweit2_data.csv')

df=data.copy()
df.loc[df['keypoint_score']<0.10,['x','y']] = np.nan

eye_dist=(((data.iloc[2][3]-data.iloc[1][3])**2)+((data.iloc[2][4]-data.iloc[1][4])**2))**(1/2)

data['x']=data['x']-data.iloc[0][3]
data['y']=data['y']-data.iloc[0][4]

data['x']=data['x']/eye_dist
data['y']=data['y']/eye_dist

data
```

#get average eye distance to standardize all values by division with that value

leftEyes=data[data['part_names']=='leftEye'].drop(['pose_score','keypoint_score'],axis=1)
leftEyes=leftEyes.rename({'x':'left_x','y':'left_y'},axis=1)

rightEyes=data[data['part_names']=='rightEye'].drop(['pose_score','keypoint_score'],axis=1)
rightEyes=rightEyes.rename({'x':'right_x','y':'right_y'},axis=1)

eyes=leftEyes.merge(rightEyes)
eyes['eye_dist']=(((eyes['left_x']-eyes['right_x'])**2)++(eyes['left_y']-eyes['right_y'])**2)**(1/2)

avg_eye_dist=eyes[eyes['eye_dist']!=0]['eye_dist'].mean()

```python
#reformat coordinates so that nose coordinate is point of origin

df_nose = (df.loc[df['part_names'] == 'nose', ['x', 'y']].copy())
df_nose.columns = ['x_nose', 'y_nose']
df = df.merge(df_nose, how='left')
df['x'] -= df['x_nose']
df['y'] -= df['y_nose']
df = df.drop(['x_nose','y_nose'], axis=1)
df.rename(columns={"x": "x_dist_to_nose", "y": "y_dist_to_nose"},inplace=True)
```

```python
#standardise all values

df['x_dist_to_nose']=df['x_dist_to_nose']/avg_eye_dist
df['y_dist_to_nose']=df['y_dist_to_nose']/avg_eye_dist
```
